implementation of kmeans algorithm using a sample data from glob datasets

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
k=5
```

```python
color=["green","red","blue","yellow","orange"]
```

```python
clusters={}
```

```python
for i in range(k):
    center=10*(2*np.random.random((x.shape[1],))-1)
    points=[]
    cluster={
        'center':center,
        'points':points,
        'color':color[i]
    }
    clusters[i]=cluster
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
assignpointstoclusters(clusters)
```

```python
# e step
def assignpointstoclusters(clusters):
    for ix in range(x.shape[0]):
        dist=[]
        curr_x=x[ix]
        for kx in range(k):
            d=distance(curr_x,clusters[kx]['center'])
            dist.append(d)
        curr_cluster=np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)
```

```python
def plotdistance(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        print(pts.shape)
#         plotting the points
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],color="black",marker="*")
```

```python
plotdistance(clusters)
```

```python
# m step- here we update every cluster's center according to mean of
# points in it
```

```python
def updateclusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_u=pts.mean(axis=0)
            clusters[kx]['center']=new_u
            clusters[kx]['points']=[]
  
```

```python
# now after m step we again will call our estep then again
# m step and so on untill these converges
```

```python
assignpointstoclusters(clusters)
plotdistance(clusters)
updateclusters(clusters)
```
